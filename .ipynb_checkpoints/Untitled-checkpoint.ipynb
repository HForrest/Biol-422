```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tqdm.notebook import tqdm
from sklearn.linear_model import LogisticRegression
%matplotlib inline
plt.style.use('ggplot')
```

```python
#Initial cleanup
df = pd.read_csv("watch_data.csv")
df.drop("Line", axis = 1, inplace= True)
df.dropna(inplace = True)
df["Timestamp"] = pd.to_datetime(df.Date + ' ' + df.Time)
df
```

```python
log_reg = LogisticRegression()
X = df.Activity.to_numpy().reshape(-1,1)
y = df["Sleep/Wake"]
log_reg.fit(X,y)
```

```python
df["act_log"] = log_reg.predict_proba(X)
df.head()
```

```python
df["Date"] = df.Timestamp.transform(func = lambda x : x.day_of_year)
df['Time'] = df.Timestamp.transform(func = lambda x : x.hour + x.minute/60 + x.second/3600)
```

```python
m = df.Activity.mean()
std = df.Activity.std()
df["act_norm"] = df.Activity.transform(func = lambda x: (x-m)/std)
```
