```python
#Initial cleanup
df = pd.read_csv("watch_data.csv")
df.drop("Line", axis = 1, inplace= True)
df.dropna(inplace = True)
df["Timestamp"] = pd.to_datetime(df.Date + ' ' + df.Time)
df
```

```python
df["Date"] = df.Timestamp.transform(func = lambda x : x.day_of_year)
df['Time'] = df.Timestamp.transform(func = lambda x : x.hour + x.minute/60 + x.second/3600)
```

```python
#Minmax feature scaling of Activivty
mn = df.Activity.min()
mx = df.Activity.max()
df["act_norm"] = df.Activity.transform(func = lambda x: (x-mn)/(mx-mn))
```
